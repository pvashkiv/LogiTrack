```python
import pandas as pd

df = pd.read_excel("Loads 2024.xlsx", sheet_name="DATA")

df["start"] = pd.to_datetime(df["start"], errors="coerce")
df["end"] = pd.to_datetime(df["end"], errors="coerce")

df["Hours"] = (df["end"] - df["start"]).dt.total_seconds() / 3600
df["Days"] = (df["end"] - df["start"]).dt.total_seconds() / (3600 * 24)
df["Miles per Hour"] = df["Miles"] / df["Hours"]

print("Таблиця з розрахунками тривалості рейсів і середньої швидкості доставки (в милях за годину)")
df[["start", "end", "From City", "To City", "Miles", "Hours", "Days", "Miles per Hour"]].head(100000)
```
